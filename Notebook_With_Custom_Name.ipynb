# OECD Vaccination rates

```python
import pandas as pd
from plot import make_plot
```

```python
url = 'http://www.oecd.org/els/family/CO_1_4_Childhood_vaccination.xlsx'
df = pd.read_excel(url, sheet_name="VaccinationRates", skiprows=3)
df.loc[1::2, 'Country'] = df.loc[::2]['Country'].values
```

```python
dfDTP = df.iloc[::2, :]
dfM = df.iloc[1::2, :]

dfDTP = dfDTP[dfDTP[2017] != '..']
dfM = dfM[dfM[2017] != '..']
```

## Percentages for Measles

```python
make_plot(dfM, url)
```

## Percentages for Diphtheria, tetanus and pertussis

```python
make_plot(dfDTP, url)
```
